#Analysing movie duration and ratings
This dataset was downloaded from kaggle.com in 2022/10/17 so I could do a study on movie durations and ratings


```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import re #regular expressions
import matplotlib.pyplot as plt #graphing
```

```python
#making a pandas dataset from the .csv input
dataset_path = '/content/drive/MyDrive/Dataset/Top_1000_Highest_Grossing_Movies_Of_All_Time.csv'
df = pd.read_csv(dataset_path)
```

```python
#sorting the dataframe by the duration of the movies
df = df.sort_values(by = 'Duration', ascending = False)
```

```python
#plotting the 10 longest movies in the dataset
plt.bar(df['Movie Title'].iloc[:10], df['Duration'].iloc[:10])
plt.xticks(rotation = 90)
plt.show()
```

```python
#plotting the graph showing the relation between duration and rating
plt.scatter(df['Duration'], df['Movie Rating'], s=12) #all movies duration/rating
plt.scatter(np.mean(df['Duration']), np.mean(df['Movie Rating'])) #mean movie duration/rating
plt.xlabel('Movie duration')
plt.ylabel('Movie rating')
plt.show()
```
